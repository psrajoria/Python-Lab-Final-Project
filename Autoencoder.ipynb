```python
import random

def random_dna_sequence(length):
    return ''.join(random.choice('ACTG') for _ in range(length))

def base_frequency(dna):
    d = {}
    for base in 'ATCG':
        d[base] = dna.count(base)/float(len(dna))
    return d

for _ in range(20):
    dna = random_dna_sequence(100)
    print(dna, base_frequency(dna))
```
