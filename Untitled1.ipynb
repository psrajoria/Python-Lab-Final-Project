```python
import sys
import string
import random
import numpy as np
import pandas as pd
from burrows import Burrows
```

```python
def generate_dataset(csv_name,rows):
    pure_str = []
    bwt_list = []
    for i in range(0,rows):
        genome = ''.join(random.choice(string.ascii_uppercase) for _ in range(random.randint(6,8)))
        b = Burrows(genome)
        bwt_str = b.bwt()
        pure_str.append(genome)
        bwt_list.append(bwt_str)

    df = pd.DataFrame()
    df['String'] = pure_str
    df['BWT_string'] = bwt_list
    if '.csv' not in csv_name:
        csv_name = f'{csv_name}.csv'
    df.to_csv(csv_name,index=False)
```

```python
if __name__ == '__main__':
    name = sys.argv[1]
    rows = int(sys.argv[2])
    generate_dataset(name,rows)
```

```python
import random
import string

def get_random_string(length):
    # choose from all lowercase letter
    letters = string.ascii_lowercase
    result_str = ''.join(random.choice(letters) for i in range(length))
    return result_str.upper()
```

```python
res = [random.randrange(5, 20, 1) for i in range(200)]
res
```

```python
genome_string = []
for i in res:
    genome_string.append( get_random_string(i) )

genome_string = np.array(genome_string)
genome_string
```

```python
rand = [random.randrange(0, len(genome_string[i])-3, 1) for i in range(200)]
rand
```

```python
res_target = [random.randrange(rand[i], len(genome_string[i])-1, 1) for i in range(200)]
res_target
```

```python
target = []
for i in range(200):
    target.append(genome_string[i][:res_target[i]])
target
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_genome = LabelEncoder()
encoded_string = labelencoder_genome.fit_transform(genome_string)
encoded_string
```

```python
encoded_target = labelencoder_genome.fit_transform(target)
encoded_target.shape
```

```python
X = []
for i in range(0, len(genome_string)):
    X.append([ genome_string[i], encoded_string[i] ])
X = np.array(X)
X
```

```python
y = []
for i in range(0, len(target)):
    y.append([ target[i], encoded_target[i] ])
y = np.array(y)
y.shape
```

```python
cat = OneHotEncoder()
one_hot_X = cat.fit_transform(X).toarray()
one_hot_X.shape
```

```python
cat = OneHotEncoder()
one_hot_y = cat.fit_transform(y).toarray()
one_hot_y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(one_hot_X, one_hot_y, test_size = 0.2, random_state =4) 
```

```python
y_test[1]
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
import matplotlib.pyplot as plt
model = Sequential()
```

```python
model.add(LSTM((1),batch_input_shape=(None, None, 1), return_sequences=False))
```

```python
model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_test,y_test))
```

```python
results = model.predict(X_test)
results
```

```python
plt.scatter(range(40),results,c='r')
plt.scatter(range(40),y_test,c='g')
plt.show()
```
